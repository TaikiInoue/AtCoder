<h3>問題文</h3> 
<section> 
高橋君のたこ焼き屋で使っているたこ焼き器は焼く場所によって美味しさの変わるクセの強いたこ焼き器です。<br> 
また、店員の力量によって一度に焼けるたこ焼きの数が違います。<br> 
高橋君はそれぞれの店員ができるだけ美味しくたこ焼きを焼けるようにしようと思いました。<br><br>

それぞれの店員は一度に焼けるたこ焼きの上限<var>\(P<sub>k</sub>\)</var>が決まっています。<br> 
また、一度に焼くたこ焼きは必ずたこ焼き器の長方形の部分になっていて、その中の全てを使わなければなりません。<br> 
それぞれの店員について一度に焼けるたこ焼きの美味しさの合計の最大値を求めて下さい。<br> 
ただし、店員が焼き始める時はたこ焼き器が完全に空いていてどの場所でも使えるとします。
