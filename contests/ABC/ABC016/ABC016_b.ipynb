<h3>問題文</h3> 
 
<p>高橋くんはプログラミングコンテストで <code>A+B</code> を出力するプログラムを書いたつもりでしたが、 <code>A-B</code> を出力するプログラムを書いたような気がしてきて不安になりました。高橋くんは書いたプログラムをうっかり消してしまって入出力しか残っていません。高橋くんが書いたプログラムが <code>A+B</code> なのか、 <code>A-B</code> なのか、どちらの可能性もあるか、どちらでもないかを判定してください。</p> 
<p>ただし、 <code>A+B</code> と <code>A-B</code> のどちらかの可能性しかない場合、そのプログラムを書いたと判定します。例えば <code>1 1</code> という入力に対して <code>2</code> を出力しているなら、 <code>A+B</code> の可能性はあるが <code>A-B</code> の可能性はないので、高橋くんは <code>A+B</code> を書いたと判定してください。</p>
