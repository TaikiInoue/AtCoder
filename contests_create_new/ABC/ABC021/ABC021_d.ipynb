<h3>問題文</h3> 
 
<p>新入社員の高橋君は、とある企業の新人プログラマーとして部署に配属されました。 
高橋君が担当した初めての仕事は、以下の擬似コードで表されるプログラムを高速化するというものでした。</p> 
 
<pre>n←(標準入力) 
ans←0 
for i=1..n 
  for j=i..n 
    ans ← ans+1 
ansの値を表示 
</pre> 
 
<p>高橋君にかかってしまえばこんな仕事はお茶の子さいさいです。

<p>劇的な高速化に成功した高橋君への部署からの期待は鰻登りです。そこで、上司は彼に更なる仕事を与えることにしました。</p>

  for a_2=a_1..n 
    for a_3=a_2..n 
      … 
      for a_k=a_{k-1}..n // a_0=1とする

</pre>
