<section> 
<h3>問題文</h3> 
 
<p>次のような制約を満たす数式 <var><span class='MathJax_Preview' style='color: inherit;'><span class='MJXp-math' id='MJXp-Span-1'><span class='MJXp-mi MJXp-italic' id='MJXp-Span-2'>S</span></span></span><script type='math/tex' id='MathJax-Element-1'>S</script></var> が与えられます。</p> 
 
<ul class='simple'> 
 
<li>演算子は <code>+</code> (加算) と <code>*</code> (乗算) のみからなる。乗算を優先して計算する。</li> 
 
<li>括弧は存在しない。</li> 
 
<li>それぞれの項は、 <var><span class='MathJax_Preview' style='color: inherit;'><span class='MJXp-math' id='MJXp-Span-3'><span class='MJXp-mn' id='MJXp-Span-4'>1</span></span></span><script type='math/tex' id='MathJax-Element-2'>1</script></var> 桁の整数である。</li> 
 
</ul> 
 
<p>例えば、<code>1+3*4*0</code> 、 <code>1+2+3+4+5</code> などの数式はこの条件を満たしますが、<code>12+3+5</code>、<code>4*6*7-3</code>、<code>(3+4)*5+2</code> のような数式は 
条件を満たさないため、入力として与えられません。</p> 
 
<p>あなたは、この数式のうち数字の部分をいくつか選んで <code>0</code> に書き換えることで、この式の値を <var><span class='MathJax_Preview' style='color: inherit;'><span class='MJXp-math' id='MJXp-Span-5'><span class='MJXp-mn' id='MJXp-Span-6'>0</span></span></span><script type='math/tex' id='MathJax-Element-3'>0</script></var> にしたいです。式の値を <var><span class='MathJax_Preview' style='color: inherit;'><span class='MJXp-math' id='MJXp-Span-7'><span class='MJXp-mn' id='MJXp-Span-8'>0</span></span></span><script type='math/tex' id='MathJax-Element-4'>0</script></var> にするために 
<code>0</code> に書き換えなければならない数字の個数の最小値を求めてください。</p> 
 
</section>
