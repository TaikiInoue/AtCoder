<p>この問題には2つのデータセットがあり，データセット毎に部分点が設定されている．</p>

<li><var><span class='MathJax_Preview' style='color: inherit;'><span class='MJXp-math' id='MJXp-Span-182'><span class='MJXp-mi MJXp-italic' id='MJXp-Span-183'>Q</span><span class='MJXp-mo' id='MJXp-Span-184' style='margin-left: 0.333em; margin-right: 0.333em;'>=</span><span class='MJXp-mn' id='MJXp-Span-185'>1</span></span></span><script type='math/tex' id='MathJax-Element-36'>Q=1</script></var> を満たすデータセット 1 に正解した場合は <var><span class='MathJax_Preview' style='color: inherit;'><span class='MJXp-math' id='MJXp-Span-186'><span class='MJXp-mn' id='MJXp-Span-187'>30</span></span></span><script type='math/tex' id='MathJax-Element-37'>30</script></var> 点が与えられる．</li>
