<h3>問題文</h3> 
 
<p>高橋君は AtCoder 社のマスコットキャラクターである。</p> 
 
<p>マスコットキャラクターとしての魅力を維持するために、高橋君は適度な運動をすることになっている。高橋君は週の運動時間が <var><span class='MathJax_Preview' style='color: inherit;'><span class='MJXp-math' id='MJXp-Span-1'><span class='MJXp-mi MJXp-italic' id='MJXp-Span-2'>L</span></span></span><script type='math/tex' id='MathJax-Element-1'>L</script></var> 分以上 <var><span class='MathJax_Preview' style='color: inherit;'><span class='MJXp-math' id='MJXp-Span-3'><span class='MJXp-mi MJXp-italic' id='MJXp-Span-4'>H</span></span></span><script type='math/tex' id='MathJax-Element-2'>H</script></var> 分以下でなければならない。</p>
