<h3>問題文</h3> 
 
<p>高橋君は、ある日不思議な生物を見た。</p> 
 
<p>その生物は choku 語という言語を用いていることがわかった。</p>

<ul> 
 
<li>文字列 <var><span class='MathJax_Preview' style='color: inherit;'><span class='MJXp-math' id='MJXp-Span-5'><span class='MJXp-mi MJXp-italic' id='MJXp-Span-6'>S</span></span></span><script type='math/tex' id='MathJax-Element-3'>S</script></var> が空文字列であるとき。</li> 
 
<li>文字列 <var><span class='MathJax_Preview' style='color: inherit;'><span class='MJXp-math' id='MJXp-Span-7'><span class='MJXp-mi MJXp-italic' id='MJXp-Span-8'>S</span></span></span><script type='math/tex' id='MathJax-Element-4'>S</script></var> が、choku 語である文字列 T の末尾に <code>ch</code> をつけた文字列であるとき。</li>
