<h3>問題文</h3> 
 
<p>高橋くんは、ソフトウエアが期待通りに動いたというエビデンス（証拠）として、画面のスクリーンショットを表計算ソフトに貼り付ける作業を命じられました。

また、表計算ソフトに貼りつけ可能なスクリーンショットには <var><span class='MathJax_Preview' style='color: inherit;'><span class='MJXp-math' id='MJXp-Span-3'><span class='MJXp-mn' id='MJXp-Span-4'>2</span></span></span><script type='math/tex' id='MathJax-Element-2'>2</script></var> つの制約が存在します。</p> 
 
<ol class='arabic simple'> 
 
<li>表計算ソフトの幅は <var><span class='MathJax_Preview' style='color: inherit;'><span class='MJXp-math' id='MJXp-Span-5'><span class='MJXp-mi MJXp-italic' id='MJXp-Span-6'>W</span></span></span><script type='math/tex' id='MathJax-Element-3'>W</script></var> しかない。そのため、貼りつけるスクリーンショットの幅の合計値は <var><span class='MathJax_Preview' style='color: inherit;'><span class='MJXp-math' id='MJXp-Span-7'><span class='MJXp-mi MJXp-italic' id='MJXp-Span-8'>W</span></span></span><script type='math/tex' id='MathJax-Element-4'>W</script></var> 以下でなければならない。</li> 
 
<li>表計算ソフトは <var><span class='MathJax_Preview' style='color: inherit;'><span class='MJXp-math' id='MJXp-Span-9'><span class='MJXp-mi MJXp-italic' id='MJXp-Span-10'>K</span></span></span><script type='math/tex' id='MathJax-Element-5'>K</script></var> 枚より多くのスクリーンショットを貼りつけることが出来ない。よって、表計算ソフトに貼りつけ可能なスクリーンショットは <var><span class='MathJax_Preview' style='color: inherit;'><span class='MJXp-math' id='MJXp-Span-11'><span class='MJXp-mi MJXp-italic' id='MJXp-Span-12'>K</span></span></span><script type='math/tex' id='MathJax-Element-6'>K</script></var> 枚までである。</li> 
 
</ol> 
 
<p>さらに、スクリーンショットには「重要度」が存在するため、高橋くんは <var><span class='MathJax_Preview' style='color: inherit;'><span class='MJXp-math' id='MJXp-Span-13'><span class='MJXp-mn' id='MJXp-Span-14'>2</span></span></span><script type='math/tex' id='MathJax-Element-7'>2</script></var> つの制約を満たしながら、貼り付けるスクリーンショットが持つ重要度の合計値を最大化したいです。 
しかし、彼にとってこの仕事は難しいので、あなたが彼の代わりに表計算ソフトに貼り付け可能なスクリーンショットが持つ重要度の合計の最大値を求めてください。</p>
