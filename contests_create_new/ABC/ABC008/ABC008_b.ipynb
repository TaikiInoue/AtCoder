<section> 
<h3>問題文</h3> 
 
<p>とある組織で、リーダーを選ぶ選挙が行われた。</p> 
 
<p>組織は <var><span class='MathJax_Preview' style='color: inherit;'><span class='MJXp-math' id='MJXp-Span-1'><span class='MJXp-mi MJXp-italic' id='MJXp-Span-2'>N</span></span></span><script type='math/tex' id='MathJax-Element-1'>N</script></var> 人の構成員で構成されており、各人は最もリーダーにふさわしい人物の名前を書いた。</p> 
 
<p>リーダーは、得票数が最も多い人物が選ばれることになっている。</p> 
 
<p>得票数が最も多い人物の名前を出力せよ。得票数が最も多い人物が複数いる場合は、そのうちどの名前を出力してもよい。</p>
