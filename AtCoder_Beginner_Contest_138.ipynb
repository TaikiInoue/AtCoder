与えられるグラフは木である。

```python
n, q = map(int, input().split())

links = [[] for _ in range(n)]
for _ in range(n - 1):
    a, b = map(int, input().split())
    links[a - 1].append(b - 1)
    links[b - 1].append(a - 1)  

values = [0] * n
for _ in range(q):
    p, x = map(int, input().split())
    values[p - 1] += x

q = [(0, -1)]
while q:
    me, parent = q.pop()
    for child in links[me]:
        if child == parent:
            continue
        values[child] += values[me]
        q.append((child, me))

print(*values)
```
